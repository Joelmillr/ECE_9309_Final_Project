```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exp1 = pd.read_csv('Experiment_1_Reward_Data.csv')
```

```python
# Set the style of the plot
plt.style.use('ggplot')

# Plot the dataframe
exp1.plot(x='Step', y='Value', legend=False)

# size of the plot
plt.rcParams["figure.figsize"] = (20, 4)

# y axis starts at -900 and ends at -1100
plt.ylim(-900, -1100)

# Customize the plot
plt.xlabel('Step', fontsize=22)
plt.ylabel('Value', fontsize=22)
plt.title("Experiment 1: Value over time", fontsize=32)
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.show()
```
